```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
data = pd.read_csv('/content/Afiliados_activos_DM_SIS.csv')
data.head()
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# Lista de columnas a transformar
columns_to_transform = ['CON_DX_OBESIDAD', 'CON_DX_HIPERTENSION', 'CON_DX_SALUDMENTAL']

# Reemplazar valores y convertir a entero
for column in columns_to_transform:
    data[column] = data[column].apply(lambda x: 1 if x == 'SI' else 0).astype(int)

# Verifica las primeras filas para confirmar la transformación
print(data[columns_to_transform].head())
```

```python
# Eliminamos las columnas referidas por falta de datos
data.drop(['UBIGEO','UBIGEO_ULT_ATE', 'DEPARTAMENTO_ULT_ATE', 'PROVINCIA_ULT_ATE', 'DISTRITO_ULT_ATE', 'NIVEL_ULT_ATE','FECHA_CORTE'], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
# Lista de columnas de fechas
date_columns = ['FECHA_AFILIADOS_SIS', 'FECHA_PRIMER_DX']

# Convertir a datetime y formatear
for col in date_columns:
    data[col] = pd.to_datetime(data[col], format='%Y%m%d', errors='coerce')  # Asumiendo formato AAAAMMDD

```

```python
data.head()
```

```python
data.nunique()
```

```python
#Considero que actualmente el CODIGO_ANONIMIZADO no nos serviriria actualmente, lo eliminaria
data.drop(['CODIGO_ANONIMIZADO'], axis=1, inplace=True)
```

```python
data
```

```python
data.dtypes
```

```python
def dummies(x):
  dummy = pd.get_dummies(data[x])
  data.drop(columns = x , inplace = True)
  return pd.concat([data, dummy], axis = 1)
```

```python
data = dummies('TIPO_DIABETES')
data
```

```python
# Cambiar nombres de forma individual
data = data.rename(columns={'Diabetes mellitus  no especificada': 'D_n_esp', 'Diabetes mellitus asociada con desnutrición': 'D_a_d','Diabetes mellitus tipo 1': 'D_t1','Diabetes mellitus tipo 2': 'D_t2', 'Otras Diabetes mellitus especificada': 'O_D_n_esp'})
```

```python
data
```

```python
print(data.columns)
```

```python
# Crear un nuevo DataFrame solo con las columnas booleanas
df_numerico = data[['D_n_esp', 'D_a_d','D_t1','D_t2','O_D_n_esp']].astype(int)

# Combinar el DataFrame original con el nuevo DataFrame de columnas numéricas
df_final = pd.concat([df_numerico, data.drop(['D_n_esp', 'D_a_d','D_t1','D_t2','O_D_n_esp'], axis=1)], axis=1)
```

```python
df_final
```

```python
# Filtrar columnas con tipo de dato entero
data_numerica = df_final.select_dtypes(include=['number'])
(data_numerica)
```
